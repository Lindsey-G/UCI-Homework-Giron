```python
# import pandas
import pandas as pd
```

```python
# create path to csv
path = "city_data.csv"

# read in csv
city_data_df = pd.read_csv(path)

# display city_data_df
city_data_df.head()
```

```python
# needed to correct Unnamed: 0 columnn
city_id_data = city_data_df.rename(columns = {"Unnamed: 0": "City ID"})
city_id_data.head()
```

```python
# save clean csv
city_id_data.to_csv('city_id_data.csv')
```

```python
# create new path to clean csv
clean_path = "city_id_data.csv"

# read in clean_path
renamed_city_data = pd.read_csv(clean_path)

# create df to html
city_data_table = renamed_city_data.to_html('data.html')
city_data_table
```
